```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# 正規分布を描画
x = np.linspace(-5, 5, 1000)
y = stats.norm.pdf(x)
plt.plot(x, y, label="Standard Normal", color="black")

```

```python
# 平均170, 標準偏差10の正規分布
x2 = np.linspace(140, 200, 1000)
y2 = stats.norm.pdf(x2, 170, 10)
plt.plot(x2, y2, label="N(170,10^2)", color="blue")
```

```python
# t分布を描画
x = np.linspace(-5, 5, 1000)
plt.plot(x, stats.t.pdf(x, 5), label="t(df=5)", color="red")
plt.plot(x, stats.t.pdf(x, 10), label="t(df=10)", color="blue")
plt.plot(x, stats.t.pdf(x, 1000), label="t(df=1000)", color="green")
plt.plot(x, stats.t.pdf(x, 10000), label="t(df=10000)", color="orange")
plt.plot(x, stats.norm.pdf(x), label="Standard Normal", color="black", linestyle="dashed")

# y軸の範囲を統一
plt.ylim(0, 0.4)
plt.xlabel("x")
plt.ylabel("Density")
plt.legend()
plt.show()
```
